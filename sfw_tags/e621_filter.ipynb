```python
import pandas as pd

# Replace 'your_file.csv' with the actual file name and path if it's not in the current directory
df = pd.read_csv('posts-2023-11-03.csv')
```

```python
df
```

```python
filtered_df = df[(df['up_score'] >= 50) & (df['down_score'] >= -(df['up_score'] / 10)) & (df['is_deleted'] == 'f') & (df['is_pending'] == 'f') & (df['is_flagged'] == 'f') & (df['rating'] == 's')]
```

```python
filtered_df
```

```python
from collections import Counter

# Initialize a Counter object to count tags
tag_counter = Counter()

# Iterate through each row in the filtered DataFrame
for index, row in filtered_df.iterrows():
    # Split the tag_string by whitespace and update the counter
    tags = row['tag_string'].split()
    tag_counter.update(tags)
```

```python
tag_counter
```

```python
for tag, count in tag_counter.items():
    if "vagina" in tag:
        print(f"Tag: {tag}, Count: {count}")
```

```python
def good_tag(tag):
    if "imminent" in tag:
        return False
    if "tentacle" in tag:
        return False
    if "urine" in tag:
        return False
    if "sex" in tag:
        return False
    if "fuck" in tag:
        return False
    if "ahegao" in tag:
        return False
    return True

tag_counter = {tag: count for tag, count in tag_counter.items() if good_tag(tag) and count > 5}
```

```python
df_e621 = pd.read_csv('e621.csv')
df_e621
```

```python
tags_to_include = list(tag_counter.keys())
filtered_df_e621 = df_e621[df_e621['name'].isin(tags_to_include)]
```
